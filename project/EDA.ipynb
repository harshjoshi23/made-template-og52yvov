 Function to execute a query and return a DataFrame


```python

def query_db(query):
    return pd.read_sql_query(query, conn)
```

```python
# Fetching data from facilities
detail_query = """
SELECT 
    facility_name, 
    facility_state, 
    total_inmate_cases, 
    total_inmate_deaths,
    facility_type
FROM facilities
ORDER BY total_inmate_cases DESC
"""
detailed_data = query_db(detail_query)

# Top 5 facilities by number of cases
top_facilities = detailed_data.head(5)
print("Top 5 Facilities by Number of Cases:")
display(top_facilities[['facility_name', 'total_inmate_cases', 'total_inmate_deaths','facility_state']])
```

```python
# Basic statistics
print("Basic Statistics:")
display(detailed_data.describe())

```

For Facilities : 

total cases vs total deaths for top 5 facilities


```python
plt.figure(figsize=(14, 7))
sns.barplot(x='facility_name', y='total_inmate_cases', data=top_facilities, label='Total Cases', color='blue')
sns.barplot(x='facility_name', y='total_inmate_deaths', data=top_facilities, label='Total Deaths', color='red')
plt.title('Top 5 Facilities: Cases vs Deaths')
plt.ylabel('Counts')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

state-wise distribution of cases and deaths


```python

# State-wise aggregation of cases and deaths
state_wise = detailed_data.groupby('facility_state').sum().reset_index()
state_wise.sort_values(by='total_inmate_cases', ascending=False, inplace=True)
```

```python
plt.figure(figsize=(16, 8))
sns.barplot(x='facility_state', y='total_inmate_cases', data=state_wise, color='blue', label='Total Cases')
sns.barplot(x='facility_state', y='total_inmate_deaths', data=state_wise, color='red', label='Total Deaths')
plt.title('State-wise Total Cases and Deaths')
plt.xticks(rotation=90)
plt.ylabel('Counts')
plt.legend()
plt.show()
```
